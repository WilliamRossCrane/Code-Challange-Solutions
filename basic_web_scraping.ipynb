```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Use a simple site for learning
url = 'http://quotes.toscrape.com/'

# Send request
response = requests.get(url)

print("Status code:", response.status_code)

# Parse
soup = BeautifulSoup(response.text, 'html.parser')

# Find all quote texts
quotes = soup.find_all('span', class_='text')

# Convert to list
quote_list = [q.get_text(strip=True) for q in quotes]

# Show the list
print(quote_list)
```

```python
df = pd.DataFrame({
    'Quote': quote_list
})

# See it nicely
df.head()
```
